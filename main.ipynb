# g3_5.py
## 3.5 Performance Comparison Using Diebold and Mariano (DM) Test
* Conduct the Diebold and Mariano (1995) test for equal predictive ability to compare the predictive model forecasts against the mean benchmark forecasts.

**In this test the H0 (NULL) Hypothesis consists of d equal to zero.
The alternative hypothesis comes from d different from zero.**

*Where d is defined as the difference from the residuals of the two variables compared (in our case the mean benchmark and a predictor varible).
So, $d_i= res_a^2 - res_b^2$*

The test carried out is a **billateral test** and the chose confidence value picked was **95%** resulting in an alpha equal to 2.5% for rejection.

The results from the comparative analysis of the optimal portfolios based on the mean benchmark forecast and predictive model forecasts appear reasonable within the context of financial analysis. Here's an interpretation of the results:

1. **Mean Return**: This indicates the average expected return of each model. The models show a range of mean returns, with 'Forecast_E12' showing the lowest and 'Combined_Forecast' the highest. This variation suggests that different predictive models lead to different portfolio compositions, impacting expected returns.

2. **Volatility**: This measures the risk or uncertainty associated with each model's returns. The 'Forecast_E12' model shows significantly higher volatility, indicating a riskier portfolio. In contrast, other models, including the benchmark, have lower and similar levels of volatility, suggesting a more conservative risk profile.

3. **Sharpe Ratio**: This ratio evaluates the return of an investment compared to its risk. Higher values are better, indicating higher returns for the same level of risk. Most models, including the benchmark, have Sharpe Ratios above 1, suggesting they offer a good risk-return trade-off. The 'Forecast_E12' model, despite its higher risk (volatility), does not offer a proportionally higher return, resulting in a lower Sharpe Ratio.

### Economic Gains/Significance

Regarding the question of whether statistical evidence of out-of-sample forecasting performance translates into economic gains or significance:

- **Performance vs. Benchmark**: The models 'Forecast_b/m', 'Forecast_tbl', 'Forecast_ntis', 'Forecast_infl', and the 'Combined_Forecast' all outperform the benchmark in terms of Sharpe Ratio, implying better risk-adjusted returns. This suggests that these models might provide economic gains over the benchmark strategy.

- **Risk-Return Trade-off**: While some models show higher mean returns than the benchmark, they do so with different levels of risk. For instance, 'Forecast_E12' has a much higher volatility, which might not be suitable for all investors despite its higher return.

- **Model Specificity**: The economic significance of out-of-sample forecasting performance heavily depends on the specific model used. For example, models with a higher Sharpe Ratio than the benchmark may offer economic gains, but this is contingent upon the reliability and consistency of the model's predictions.

- **Practical Considerations**: In real-world scenarios, factors like transaction costs, market impact, and tax implications can significantly affect the economic gains from any investment strategy. These factors need to be considered alongside the statistical evidence.

In conclusion, while the statistical evidence suggests that some models offer better risk-adjusted returns than the benchmark, whether this translates into economic gains depends on the investor's risk appetite, investment goals, and practical considerations in implementing the strategy.
